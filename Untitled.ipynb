```python
import pandas as pd
dag_sizes = [25, 50, 100, 200, 400]
```

```python
all_sheets = {}
for dag in dag_sizes:
    filename = f'./results/DAG{dag}/EDLS.xlsx'
    sheets = []
    for i in range(4):
        sheets.append(pd.read_excel(filename, sheet_name=i))    
    all_sheets[dag] = sheets
```

```python
def plot(dag_size, col_name, sheet_num):
    data = all_sheets[dag_size][sheet_num]
    data.plot.line(x='Processor Combination', y=col_name)
    
```

```python
plot(25, 'Total Energy', 0)
```

```python
avg_exec_times = [sheet['Total Execution Time'].mean() for sheet in sheets]
avg_energy = [sheet['Idle Energy'].mean() for sheet in sheets]
```
